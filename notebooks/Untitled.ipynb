```python
# The .gz file can be read directly by pandas - no need to extract it!
filepath = r"C:\Users\DELL G3\Desktop\RMIT SEM - 4\Loan Club Project\data\accepted_2007_to_2018Q4.csv.gz"

print("Loading data... this will take 1-2 minutes due to file size, please wait.")

df = pd.read_csv(filepath, compression='gzip', low_memory=False)

print(f"Done! Dataset shape: {df.shape}")
print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
```

```python
# How much data is missing per column?
missing = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing %': (df.isnull().sum() / len(df) * 100).round(2)
})

# Only show columns that actually have missing values
missing = missing[missing['Missing Count'] > 0].sort_values('Missing %', ascending=False)

print(f"Columns with missing values: {len(missing)} out of {df.shape[1]}")
print("\n", missing.head(30))
```
